### Assigning to arrays

```python
import numpy as np
A=np.arange(1,28).reshape(3,3,3)
```

```python
# You can assign an element
A[0][0][0]=-10
A[0,1,2] = -100
print(A)
```

```python
# You can also assign a subarray by slicing

# The value can be an array of the same shape
A[0,:,:]=np.arange(-9,0).reshape(3,3) # Equivalent to A[0]
A[1]=np.arange(-18,-9).reshape(3,3)
print(A)
```

```python
# The value can be a constant
A[1,1,:]=-50
print(A)
print('*'*10)
```

```python
print(A[1])
print(A[0])
print(A[ [0,1] ])
print('-'*50)
print(A)
```

```python
# The value can also be a subarray that can be broadcasted
A=np.arange(1,28).reshape(3,3,3)

#A[2,:]=np.arange(1,4)
#A[[1,2]]=np.arange(1,4) #A[1] = np.ara  and A[2] = np.ara
A[[1,2], [0,1], :]=np.arange(1,4)
A[[0,2],[1,2],[0,1]] = -90000
A[[0,2],[1,2],2] = -9999
print(A)
```

```python
a = np.arange(12)**2                       # the first 12 square numbers
print(a)
#i = np.array([1, 1, 3, 8, 5])              # an array of indices
#print(a[i])                                      # the elements of a at the positions i
#print(a[[3,4]])

j = np.array([[3, 4], [9, 7],[8,8],[8,8]])      # a bidimensional array of indices
print(a[j])                        # the same shape as j
print(a[j].shape)
print(a)
print(a.shape)
```

### Integer Indexing

```python
A=np.arange(1,65).reshape(4,4,4)
```

```python
# You can index an array with a list/array of indices
print(A)
print(A[[0,1]])
print("\n"+"-_"*20+"\n")

# Just a single array indexes the first dimension, same as

print(A[[0,1],:,:])
print("\n"+"-_"*20+"\n")

# You can also index the other dimensions
print(A[:,:,[0,2]])
```

```python
# You can repeat an index in your array

print(A[[0,1,0,0,1],:,:])
```

```python
# You can give a list of indexes for each dimension
print(A[[0,1,0,0],
        [1,2,2,2],
        [0,1,2,3]])
```

```python
# The list of indices can be in a more complicated shape
print(A)
i=np.array([[0,1],[1,0]])
j=np.array([[1,2],[2,2]])
k=np.array(       [1,2])  #=> [[1,2],[1,2]]
A[i,j,k]
```

```python
# Shuffle an array
A=np.random.randint(1,20,10)
print(A)
shuffle=np.random.permutation(A.shape[0]) #np.random.shuffle(np.arange(number))
print(shuffle)
print(A[shuffle]) #np.random.shuffle(A)
```

```python
# Sort an array according to another array

A=np.random.randint(1,20,10)
B=np.random.randint(1,20,10)

sorted_idcs=np.argsort(A)
B_sorted=B[sorted_idcs]
```

```python
# You can also assign using integer indices

A=np.arange(1,28).reshape(3,3,3)
print(A)
A[[1,2],[0,2],2]=-50

print(A)
```

```python
a = np.arange(12).reshape(3,4)
print(a)
print(a[[[0,1],[1,2]], [[2, 1],[3, 3]]])


i = np.array([[0, 1],[1, 2]])                     # indices for the first dim of a
              
j = np.array([[2, 1],[3, 3]])                     # indices for the second dim
              

print(a[i, j])                                   # i and j must have equal shape
print(a[i,2])
```

### Boolean Indexing

```python
A=np.random.randint(1,50,12).reshape(3,4)
print(A)
```

```python
# You can do slices or selections using boolean arrays same length as the dimension
# This is recommended to do only for one dimension

# selecting rows
#print(A[[True,True,True],:])
print("\n"+"-_"*20+"\n")

# selecting columns
#print(A[:,[True,False,False,True]])
print("\n"+"-_"*20+"\n")

# selecting both rows and columns is confusing
# and not recommended

print(A[[True,False,True],[True,False,False,True]])
```

```python
# Generally you'll use this in a more complicated way
# For example let's get the rows that have maximums > some value

A=np.random.randint(0,50,(4,6))
print(A)
print("\n"+"-_"*20+"\n")

maxes=np.max(A,axis=1)
print(maxes)
print(maxes>40)
print("\n"+"-_"*20+"\n")

print(A[maxes>40,:])

#np.where(condition, true_case, false_case)
```

```python
# You can also index by a boolean array of the size as the 
# original
# It gives a flat array with all the entries that had True
print(A>10)
print(A[A>10])
```

```python
# Again you can use Boolean indexing to assign values

A=np.random.randint(0,50,(4,6))
print(A)
print("\n"+"-_"*20+"\n")
print(A[:,0])
A[A[:,0]<30]=np.arange(-6,0)
print(A)
```

### np.inf, np.nan

```python
# Numpy has a special type to represent infinity

print(np.inf)

# Actually regular python also has infinity

print(float("inf"))
```

```python
# Let's see how np.inf behaves

# Multiplying by a negative number gives -np.inf
x=np.inf
print(x)
x*=-1
print(x)

# The opposite also works

x*=-5
print(x)
```

```python
# Adding,subtracting,dividing,multiplying with normal
# numbers gives np.inf ( or -np.inf) except when multiplying
# or dividing with a negative number

print(np.inf+5)
print(-np.inf-6)
print(-np.inf*-30)
print(-np.inf/-22)
```

```python
# Dividing a number by np.inf gives 0

print(1223/np.inf)
```

```python
# Interestingly dividing by 0 in numpy arrays gives infinity

print(np.array([222])/0)
```

```python
# What about operations between np.inf

# These have some meaning
print(np.inf + np.inf)
print(np.inf*np.inf)

# These are absolutely meaningless
print(np.inf-np.inf)
print(np.inf/np.inf)
```

```python
# Which brings us to another special numpy data type
# nan : Not a Number
print(np.nan)

# Numpy uses nan to designate invalid values
# You can use it too

print(np.array([1,1,np.nan,2]))
```

```python
# Some other ways you can get np.nan

print(np.sqrt(-2))
print(np.zeros(1)/np.zeros(1))
print(np.inf * 0)
print(np.log(-2))
```

```python
# Any operation with np.nan gives np.nan

print(1+np.nan)
print(np.inf + np.nan)
print(np.nan/np.array([0]))
print(np.nan - np.nan)
```

```python
print(100/0)
```

1. There might be some valid use cases for np.inf, but mostly you will encounter it when some calculation error happens - similar to division by zero.
2. If you see np.nan then clearly some undefined arithmetic operation has happened in your code.

## Reading data

### The main interface

```python
# Columns can be accessed using dot notation
df.birth_year
```

```python
# Columns can also be indexed - This will return a series
df["birth_year"]
```

```python
# Multiple columns can also be indexed - This will return a DataFrame
df[["first_name","birth_year"]]
```

```python
# You can slice the rows by position
sliced=df[50:100]
# Note that the index still keeps the old numbers
# Thus index itself is kind of like a series - it's not just numbering from top to bottom.
# Actually, the index can be strings, dates anything
# Though for simplicity we will stick to integer indexes
print(sliced.index)
sliced.head()
```

```python
# You can chain the two indexing style
df[:50][["first_name","last_name"]].head(5)
```

To summarise, indexing the dataframe directly like ``df[]`` allows
* indexing a single column by name (or label)
* indexing a list of columns by name 
* slicing of rows by position.

Note that the following things doesn't work
* indexing columns by position
* indexing/slicing rows by name/label

As we will see, there are two other interfaces that give us this functionality.

### .loc

```python
# Index a column by name same as before
# Notice that now you have to compulsarily provide a slice for the rows
df.loc[:,["first_name","last_name"]]
```

```python
# But let's look at rows
sliced=df[50:100]
# Note the index goes from 50 to 100
sliced.head()
```

```python
# We can index by position with the main interface - for example : 0:5 returns 50-54
sliced[0:5]
```

```python
# But can we slice with respect to the original index?

# This will return an empty DF, as position is only upto 50
sliced[55:60]
```

```python
# But this works well
sliced.loc[55:60]
```

```python
# Sometimes the original index is useful, especially if you want to modify the slice
# and assign to the original DataFrame (it automatically matches the indexes)

# But sometimes after a slice you want to reset the index
sliced=sliced.reset_index()
sliced
```

### Boolean indexing

## Setting data

```python
# The most common operation : computing a new column based on other columns

only_90s_kids["age"] = 2020 - only_90s_kids.birth_year

# Notice how the index is preserved
only_90s_kids.head(5)
```

```python
# You could even have assigned it to the original df

# but let's copy it first ;)
fresh_df=df.copy()

# All the correct indices will be filled otherwise it will be set as NaN.
# This is one of the use cases of having a separate index object
# vs just position numbering like numpy
fresh_df["age"] = 2020 - only_90s_kids.birth_year
fresh_df
```

```python
# You can of course set values at a particular position
print(fresh_df.iloc[0,-1]) 
fresh_df.iloc[0,-1]=29
print(fresh_df.iloc[0,-1]) 
```

# Data Investigation Example 1

Let's make a set of questions and then see how we can answer them by using pandas

* Who are the employers?
* What is the employee count of each employer?
* What is the % of Male vs Female employees for each company?
* What is the average age of an employee for each company?
* Which bank is most popular in each company?

```python
df[:]["employer"].drop_duplicates()
```

```python
grps=df.groupby(by="employer")
```

```python
grps.groups
```

```python
# Notice that it counts non-empty values
grps.count()
```

```python
# So how to get the number of rows in each group?

# One way:
fresh=df.copy()
fresh["id"]=np.arange(0,len(fresh))
display(fresh.groupby("employer").count()["id"])

# Another simpler way :)
display(grps.size())

# Remember a little creativity goes a long way
```

```python
grps=df.groupby(by=["employer","gender"])

# This is an example of a hierarchical index for a series
display(grps.size())
```

```python
counts=pd.DataFrame(grps.size())
print(counts.to_records())
counts=pd.DataFrame(counts.to_records())
display(counts)
```

```python
with_age=df.copy()
with_age["age"]=2020 - with_age.birth_year
with_age[["employer","age"]].groupby("employer").mean()
```

```python
temp=df[["employer","bank"]].groupby(["employer","bank"]).size()
temp=pd.DataFrame(pd.DataFrame(temp).to_records())
temp=temp.rename(columns={"0":"count"})
temp=temp.sort_values('count',ascending=False)
display(temp)
```

```python
# Some more creativity
display(temp.drop_duplicates(subset=["employer"],keep="first"))
```
